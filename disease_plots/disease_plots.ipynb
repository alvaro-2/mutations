# Disease plots from Disphase BD

```python
%load_ext autoreload
%autoreload 2
```

```python
import disease_plots
import downloads
from os.path import join
```

## Download data

```python
downloads.download_mesh_data()
```

## Set common data

```python
datafile = join(
    '../data', 'd2021.bin'
)
nodes = disease_plots.build_mesh_tree_data(
    datafile=datafile
)
mesh_uis = disease_plots.MeshUISelectorAllMutations().get_mesh_ui_by_protein()
tags_names = disease_plots.select_tag_names(
    nodes
)
```

## Disease plots - C (Disease) MeSH Category

```python
disease_hits = disease_plots.get_mesh_ui_by_disease(
    mesh_uis,
    nodes,
    min_tag_depth=1,
    max_tag_depth=1
)
hits = disease_plots.filter_hits(disease_hits, start_with=[])
disease_counts = disease_plots.count_disease_tags(hits)
disease_counts = disease_plots.filter_disease_counts(disease_counts, 5)
disease_counts = disease_plots.remove_tags_with_same_name(
    disease_counts,
    tags_names
)
disease_counts.sort(key=lambda x: x[1], reverse=False)
disease_plots.draw_plot(
    disease_counts,
    tags_names,
    figsize=(16,10),
    # outfile="top_c04.png",
)
```

## Disease plots - C04 (Disease - Neoplasms) MeSH Category

```python
disease_hits = disease_plots.get_mesh_ui_by_disease(
    mesh_uis,
    nodes,
    min_tag_depth=2,
    max_tag_depth=2
)
hits = disease_plots.filter_hits(disease_hits, start_with=['C04'])
disease_counts = disease_plots.count_disease_tags(hits)
disease_counts = disease_plots.filter_disease_counts(disease_counts, 0)
disease_counts = disease_plots.remove_tags_with_same_name(
    disease_counts,
    tags_names
)
disease_plots.draw_plot(
    disease_counts,
    tags_names,
    figsize=(16,8),
    # outfile="top_c04.png",
)
```
